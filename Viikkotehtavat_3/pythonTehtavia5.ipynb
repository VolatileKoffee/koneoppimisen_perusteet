```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from google.colab import data_table
from vega_datasets import data

# Valitse datasetti Vega Datasets -kirjastosta
df = data.cars()  # Esimerkkinä käytetään 'cars' datasettiä

# Ota käyttöön Google Colabin interaktiivinen taulukkonäkymä
data_table.enable_dataframe_formatter()

# Suodatetaan kaikki Volvon autot
volvo_autot1 = df.loc[df['Name'] == 'Volvo']

# Näytetään suodatettu data
volvo_autot1
if len(volvo_autot1) == 0:
  print("Yep, dataframe 'volvo_autot1' on tyhja.")
# --> ei tuloksia


# Suodatetaan autot, joiden nimi sisältää 'Volvo'
volvo_autot2 = df.loc[df['Name'].str.contains('Volvo', case=False, na=False)]
volvo_autot2
if len(volvo_autot2) != 0:
  print("Yep, dataframessa 'volvo_autot2' on tuloksia!")
# --> tuloksia tulee!


# Lisätään uusi sarake, joka sisältää Horsepower ja Cylinders -sarakkeiden osamäärän
df['Power_per_Cylinder'] = df['Horsepower'] / df['Cylinders']

# Näytetään päivitetty DataFrame
df

# --> uusi 'Power_per_Cylinder' sarake on lisatty

# Oletetaan, että df on alkuperäinen DataFrame
# Pudotetaan sarakkeet 'Horsepower' ja 'Cylinders' näkyvistä
df_filtered = df.drop(columns=['Year', 'Origin'])

# Näytetään suodatettu DataFrame interaktiivisessa näkymässä
df_filtered

# --> suodatettu on, sarakkeet 'Year' ja 'Origin' on poistetu

# Tehtävä 2
# Lisää df_filtered dataframeen uusi sarake, johon lasket jokaiselle autolle
# painon ja kiihtyvyyden suhteen.
df_filtered["Acceleration_per_Weight"] = df_filtered["Acceleration"] / df_filtered["Weight_in_lbs"]
df_filtered

# Selvitä tuon jälkeen mikä autotyyppi saa kilot kiihtymään nopeiten. Nimi / arvo
max_acc_per_weight_idx = df_filtered["Acceleration_per_Weight"].idxmax()
fastest_car_name = df_filtered.loc[max_acc_per_weight_idx, "Name"]
fastest_car_acc_per_weight = df_filtered.loc[max_acc_per_weight_idx, "Acceleration_per_Weight"]

print(f"Auto '{fastest_car_name}' kiihtyvyys/paino -arvolla '{fastest_car_acc_per_weight}'")

# Suodata interaktiivisesta taulukosta pois Miles_per_Gallon ja Displacement sarakkeet.
df_filtered = df_filtered.drop(columns=['Miles_per_Gallon', 'Displacement'])
df_filtered
# --> Yep, sarakeluku tippui 9-->7 (pl. index-sarake)
# tehty google colab -palvelussa
```

```python
'''
Tehtävä 3


Alla olevan esimerkkikoodin avulla saat luettua world-data-2023.csv tiedostosta
195 riviä ja 35 saraketta tietoa eri maiden tilanteesta. Tehtävät ovat seuraavat:

1. Muokkaa Pandas dataframea siten, että valitset sieltä mielenkiintoisimmat sarakkeet,
   jotka ovat:
   "Country", "Land Area(Km2)", "Armed Forces size", "Co2-Emissions" ja "Population".
      -> Vinkki: Kysy ChatGPT:ltä neuvoa, tehtävän eri vaiheissa. Sieltä saat hyviä toimivia esimerkkikoodeja käyttöösi.

2. Taulukossa on NaN lukuja, jotka ovat ongelmiallisia maksimi arvoja haettaessa. Muuta kaikki Nan => 0.
   Taulukossa on myös $ ja % merkkejä, jotka kannattaa muuttaa tyhjiksi merkeiksi ennen laskuoperaatioita.
   Taulukon numerot on esitetty muodossa 3,003,000 eli pilkut pitää muuttaa tyhjiksi merkeiksi.

3. Laske/selvitä sen jälkeen taulukosta
   a) Millä maalla on suurimmat asejoukot
   b) Millä maalla on eniten sotilaita maan kokoon verrattuna.
   c) Millä maalla on eniten sotilaita maan väestöön verrattuna.
   d) Luettele 5 suurinta Co2 päästäjää
'''

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from google.colab import data_table
from vega_datasets import data


# Ota käyttöön Google Colabin interaktiivinen taulukkonäkymä
data_table.enable_dataframe_formatter()

# asetetaan tiedosto Google Colabin files/content/sample_data/ -kansioon
filepath = "/content/sample_data/world-data-2023.csv"
original_df = pd.read_csv(filepath)

# Tehtava 1
wanted_cols = ['Country','Land Area(Km2)','Armed Forces size','Co2-Emissions','Population']
# Luetaan vain tietyt sarakkeet CSV-tiedostosta
wanted_cols_df = pd.read_csv(filepath, usecols = wanted_cols)


# Tehtava 2
# annetaan NaN-kentille arvo "0"
wanted_cols_df = wanted_cols_df.fillna(0)
# korvataan '$', '%' ja ',' merkit tyhjilla paikoilla ''
wanted_cols_df = wanted_cols_df.replace(['$','%',','],'',regex=True)

# valitaan numeromuotoon kaannettavat sarakeet
numeric_cols = ['Land Area(Km2)','Armed Forces size','Co2-Emissions','Population']
wanted_cols_df[numeric_cols] = wanted_cols_df[numeric_cols].astype(int)

# Näytetään tulos
print(f"DEBUG:\n{wanted_cols_df.head()}\n") # default int n=5

# Tehtava 3
# a)
max_armed_forces_size_country = wanted_cols_df.loc[wanted_cols_df["Armed Forces size"].idxmax(), "Country"]
print(f"Teht 3. a) Maalla {max_armed_forces_size_country} on suurimmat asejoukot.")

# b)
max_soldiers_per_country_name_idx = wanted_cols_df["Armed Forces size"] / wanted_cols_df["Land Area(Km2)"]
max_soldiers_per_country_name = wanted_cols_df.loc[max_soldiers_per_country_name_idx.idxmax(), "Country"]
print(f"Teht 3. b) Maalla {max_soldiers_per_country_name} on eniten sotilaita maan kokoon verrattuna.")

# c)
max_soldiers_per_population_name_idx = wanted_cols_df["Armed Forces size"] / wanted_cols_df["Population"]
max_soldiers_per_population_name = wanted_cols_df.loc[max_soldiers_per_population_name_idx.idxmax(), "Country"]
print(f"Teht 3. c) Maalla {max_soldiers_per_population_name} on eniten sotilaita maan vaestoon verrattuna.")

# d)
top_Co_countries = wanted_cols_df.nlargest(5, "Co2-Emissions")
print(f"Teht 3. d) Tassa ovat suurimmat Co2 paastajamaat (5 kpl):\n{top_Co_countries['Country']}")


# tehty google colab -palvelussa
```
